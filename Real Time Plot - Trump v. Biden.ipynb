```python
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation
%matplotlib notebook
```

```python
plt.style.use('fivethirtyeight')
```

```python
frame_len = 10000 # display number for tweets

fig = plt.figure(figsize=(9,16))

def animate(i):
    data = pd.read_csv('sentiment.csv')
    y1 = data['Trump']
    y2 = data['Biden']
    
    if len(y1)<=frame_len: # plot first 10000 tweets
        plt.cla()
        plt.plot(y1, label='Donald Trump', color='red')
        plt.plot(y2, label='Joe Biden', color = 'blue')
    else: # plot most recent 10000 tweets
        plt.cla()
        plt.plot(y1[-frame_len: ], label='Donald Trump', color = 'red')
        plt.plot(y2[-frame_len: ], label='Joe Biden', color = 'blue')
    
    plt.legend(loc='upper left')
    plt.tight_layout()
        
        
ani = FuncAnimation(plt.gcf(), animate, interval=100) # calling the animation function
```
